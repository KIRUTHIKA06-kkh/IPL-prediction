```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

DATA CLEANING PROCESS

```python
data = pd.read_csv('deliveries.csv')
```

```python
data
```

```python
data.isnull().sum()
```

```python
threshold = 0.5

data_cleaned = data.dropna(thresh=int((1-threshold) * len(data)), axis=1)
```

```python
data_cleaned.isnull().sum()
```

```python
data_cleaned.to_csv("ipldata.csv")
```

EXPLORATORY DATA ANALYSIS(EDA)

```python
df = pd.read_csv("ipldata.csv")
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
df
```

```python
df.shape
```

```python
df['batting_team'].unique()
```

```python
df['bowling_team'].unique()
```

```python
df['batter'].unique()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(df['batsman_runs'], kde=True, bins=20)
plt.title("Distribution of Batsman Runs")
plt.xlabel("Batsman Runs")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(df['total_runs'], kde=True, bins=20)
plt.title("Distribution of Total Runs")
plt.xlabel("Total Runs")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(df['extra_runs'], kde=True, bins=20)
plt.title("Distribution of Extra Runs")
plt.xlabel("Extra Runs")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['total_runs'])
plt.title("Boxplot of Total Runs")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['batsman_runs'])
plt.title("Boxplot of Batsman Runs")
plt.show()
```

```python
numeric_columns = df.select_dtypes(include=[np.number])
```

```python
correlation_matrix = numeric_columns.corr()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='batting_team', y='total_runs', data=df)
plt.title("Total Runs by Batting Team")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='bowling_team', y='total_runs', data=df)
plt.title("Total Runs by Bowling Team")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(x='over', y='total_runs', data=df)
plt.title("Total Runs by Over")
plt.xlabel("Over")
plt.ylabel("Total Runs")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='batting_team', data=df, order=df['batting_team'].value_counts().index)
plt.title("Frequency of Batting Teams")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='bowling_team', data=df, order=df['bowling_team'].value_counts().index)
plt.title("Frequency of Bowling Teams")
plt.xticks(rotation=90)
plt.show()
```
